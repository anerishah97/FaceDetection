```python
#import required libraries 
#import OpenCV library
import cv2
#import matplotlib library
import matplotlib.pyplot as plt
#importing time library for speed comparisons of both classifiers
import time 
%matplotlib inline
```

```python
def convertToRGB(img): 
    return cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
#load test iamge
test1 = cv2.imread('test.jpg')
```

```python
#convert the test image to gray image as opencv face detector expects gray images 
gray_img = cv2.cvtColor(test1, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(test1, cmap='gray')  
```

```python
#load cascade classifier training file for haarcascade 
haar_face_cascade = cv2.CascadeClassifier('C:\Anaconda3\Library\etc\haarcascades\haarcascade_frontalface_alt.xml')
```

```python
#let's detect multiscale (some images may be closer to camera than others) images 
# The value of scalefactor is decided after several experimentations manually
# Todo: Find a solution so that the manual adjustments are reduced.

faces = haar_face_cascade.detectMultiScale(gray_img, scaleFactor=1.006, minNeighbors=10);  
```

```python
print('Faces found: ', len(faces))
```

```python
#go over list of faces and draw them as rectangles on original colored img 
for(x, y, w, h) in faces:     
    cv2.rectangle(test1, (x, y), (x+w, y+h), (0, 255, 0), 2)
```

```python
plt.close()
#convert image to RGB and show image 
plt.imshow(convertToRGB(test1))
```
